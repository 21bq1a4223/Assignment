```python
import nltk
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.tokenize import sent_tokenize, word_tokenize

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab')

text = """I usually like listening to music while studying.
 I believe that it helps me focus, while some people feel it can be distracting.
 The real question is: does music improve concentration, or does it reduce productivity?
 Healthy habits are important to be followed by everyone.
 Eating nutritious food, exercising regularly, and getting enough sleep daily can improve us both physical and mental health.
 Doctors often say that: small lifestyle changes can make a big difference!
 """
```

```python
# Sentence Tokenization
sentences = sent_tokenize(text)
print(sentences)
```

```python
# Word Tokenization
tokens = word_tokenize(text)
print(tokens)
```

```python
# Stop words removal
stop_words = set(stopwords.words('english'))
tokens_punctuation=[w for w in sentences if w.lower() not in stop_words]
print(sentences)
```

```python
# Stemming
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in tokens_punctuation]
print(stemmed_words)
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in stemmed_words]

print("Final Preprocessed Words:", lemmatized_words)
```

```python
# Feature Extrction Process
converted_data=" ".join(lemmatized_words)
print(converted_data)
```

```python
# Bag of Words
from sklearn.feature_extraction.text import CountVectorizer
corpus=[converted_data]#corpus object assined string text data
vectorizer=CountVectorizer()
bag_of_words=vectorizer.fit_transform(corpus)
# each word freqvency how many times it appiyer in the text
print(bag_of_words.toarray())
print(vectorizer.get_feature_names_out())
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
corpus=[converted_data]
vectorizer1=TfidfVectorizer()
tfidf_vac=vectorizer1.fit_transform(corpus)
print(tfidf_vac.toarray())
print(vectorizer1.get_feature_names_out())
```

```python
# word2vec
from gensim.models import Word2Vec
model=Word2Vec(sentences,min_count=1,window=5,workers=4)
print('print vector for music')
print(model.wv['music'])
print('\nWord similar to music')
print(model.wv.most_similar('music'))
```

```python
import numpy as np
def sent_vec(sentence):
  words=sentence.lower().split()
  vectors=[]
  for word in words:
    if word in model.wv:
      vectors.append(model.wv[word])
  return np.mean(vectors,axis=0)
sentence="I usually like listening to music while studying"
vec=sent_vec(sentence)
print("sentence vector:")
print(vec)
```

```python
# Cosine Similarity
from sklearn.metrics.pairwise import cosine_similarity

s1=sent_vec("I usually like listening to music while studying")
s2=sent_vec("I believe that it helps me focus, while some people feel it can be distracting")
s3=sent_vec("Healthy habits are important to be followed by everyone")
sim_1_2 = cosine_similarity([s1], [s2])
sim_1_3 = cosine_similarity([s1], [s3])
sim_2_3 = cosine_similarity([s2], [s3])

print("Similarity between Sentence 1 & 2:", sim_1_2[0][0])
print("Similarity between Sentence 1 & 3:", sim_1_3[0][0])
print("Similarity between Sentence 2 & 3:", sim_2_3[0][0])
```

```python
# Supervised Learning
from sklearn.linear_model import LogisticRegression

X=[
    sent_vec("I usually like listening to music while studying"),
    sent_vec("I believe that it helps me focus, while some people feel it can be distracting"),
    sent_vec("Healthy habits are important to be followed by everyone"),
    sent_vec("Doctors often say that: small lifestyle changes can make a big difference"),
]
y=[1,1,0,0]
model_1=LogisticRegression()
model_1.fit(X,y)
test=sent_vec("Machine learning is a subset of AI")
print("prediction:",model_1.predict([test]))
```

```python
# Unsuperwised Learning
from sklearn.cluster import KMeans
X=[sent_vec("I usually like listening to music while studying"),
sent_vec("I believe that it helps me focus, while some people feel it can be distracting"),
sent_vec("Healthy habits are important to be followed by everyone"),
sent_vec("Doctors often say that: small lifestyle changes can make a big difference")
]
kmeans=KMeans(n_clusters=2)
kmeans.fit(X)
print("clusters",kmeans.labels_)
```
